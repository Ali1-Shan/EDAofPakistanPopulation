# About DataSet

**Description**
This dataset contains demographic information from the Pakistan Population Census conducted in 2017. It provides detailed population data at various administrative levels within Pakistan, including provinces, divisions, districts, and sub-divisions. The dataset also includes information on urban and rural populations, gender distribution, transgender individuals, sex ratios, population figures from the 1998 census, and annual growth rates.

**Features**
Province: The administrative provinces or regions of Pakistan where the census data was collected.

**Division:** The divisions within each province. Divisions are the second level of administrative divisions in Pakistan.

**District:** Districts within each division, representing larger administrative units.

**Sub-Division: **Sub-divisions or tehsils within each district, providing more localized data.

**Area:** The land area (in square kilometers) of each sub-division.

**Urban Population 2017:** The population of urban areas within each sub-division for the year 2017.

**Rural Population 2017:** The population of rural areas within each sub-division for the year 2017.

**Male Population 2017:** The male population within each sub-division for the year 2017.

**Female Population 2017:** The female population within each sub-division for the year 2017.

**Transgender Population 2017:** The population of transgender individuals within each sub-division for the year 2017.

**Sex Ratio 2017:** The sex ratio, calculated as the number of females per 1000 males, within each sub-division for the year 2017.

**Population in 1998:** The total population of each sub-division as recorded in the 1998 census.

**Annual Growth Rate:** The annual growth rate of the population in each sub-division, calculated as the percentage increase from 1998 to 2017.

**Data Source**
The data in this dataset was collected from official Pakistan Population Census reports and may include data from various government sources. It is essential to provide proper attribution and reference the original sources when using this dataset for analysis or research.

**Data Usage**

## Question 1:What is the total population in rural and urban areas for each year?

##  Question 2: How has the male-to-female sex ratio changed over time in both rural and urban areas?

````python

df['SEX RATIO 1998 (RURAL)'] = df['MALE (RURAL)'] / df['FEMALE (RURAL)']
df['SEX RATIO 1998 (URBAN)'] = df['MALE (URBAN)'] / df['FEMALE (URBAN)']


df['SEX RATIO 2017 (RURAL)'] = df['MALE (RURAL)'] / df['FEMALE (RURAL)']
df['SEX RATIO 2017 (URBAN)'] = df['MALE (URBAN)'] / df['FEMALE (URBAN)']

# Prepare data for plotting
years = ['1998', '2017']
rural_sex_ratios = [df['SEX RATIO 1998 (RURAL)'].mean(), df['SEX RATIO 2017 (RURAL)'].mean()]
urban_sex_ratios = [df['SEX RATIO 1998 (URBAN)'].mean(), df['SEX RATIO 2017 (URBAN)'].mean()]

# Plot the changes in sex ratio over time
plt.figure(figsize=(12, 6))
plt.plot(years, rural_sex_ratios, marker='o', label='Rural')
plt.plot(years, urban_sex_ratios, marker='o', label='Urban')

plt.xlabel('Year')
plt.ylabel('Sex Ratio (Male/Female)')
plt.title('Changes in Male-to-Female Sex Ratio Over Time')
plt.legend()
plt.grid(True)
plt.show()
````

````python

df['SEX RATIO 1998 (RURAL)'] = df['MALE (RURAL)'] / df['FEMALE (RURAL)']
df['SEX RATIO 1998 (URBAN)'] = df['MALE (URBAN)'] / df['FEMALE (URBAN)']

df['SEX RATIO 2017 (RURAL)'] = df['MALE (RURAL)'] / df['FEMALE (RURAL)']
df['SEX RATIO 2017 (URBAN)'] = df['MALE (URBAN)'] / df['FEMALE (URBAN)']


mean_sex_ratio_1998_rural = df['SEX RATIO 1998 (RURAL)'].mean()
mean_sex_ratio_1998_urban = df['SEX RATIO 1998 (URBAN)'].mean()
mean_sex_ratio_2017_rural = df['SEX RATIO 2017 (RURAL)'].mean()
mean_sex_ratio_2017_urban = df['SEX RATIO 2017 (URBAN)'].mean()

# Print the results
print(f"Mean Sex Ratio in 1998 (Rural): {mean_sex_ratio_1998_rural:.2f} males per female")
print(f"Mean Sex Ratio in 1998 (Urban): {mean_sex_ratio_1998_urban:.2f} males per female")
print(f"Mean Sex Ratio in 2017 (Rural): {mean_sex_ratio_2017_rural:.2f} males per female")
print(f"Mean Sex Ratio in 2017 (Urban): {mean_sex_ratio_2017_urban:.2f} males per female")
````

## Question 3:What is the average household size in rural and urban areas?

````python
avg_household_size_rural = df['AVG HOUSEHOLD SIZE (RURAL)'].mean()
avg_household_size_urban = df['AVG HOUSEHOLD SIZE (URBAN)'].mean()
print(f"Average Household Size in Rural Areas: {avg_household_size_rural:.2f}")
print(f"Average Household Size in Urban Areas: {avg_household_size_urban:.2f}")
````

## Question 4:What is the annual growth rate of the population in rural and urban areas?

````python
annual_growth_rate_rural = ((df['ALL SEXES (RURAL)'] - df['POPULATION 1998 (RURAL)']) / df['POPULATION 1998 (RURAL)']) * 100
annual_growth_rate_urban = ((df['ALL SEXES (URBAN)'] - df['POPULATION 1998 (URBAN)']) / df['POPULATION 1998 (URBAN)']) * 100


print("Annual Growth Rate in Rural Areas:")
print(annual_growth_rate_rural)
print("\nAnnual Growth Rate in Urban Areas:")
print(annual_growth_rate_urban)
````

## Question 5:How has the population distribution changed from 1998 to 2017 in rural and urban areas?

````python
rural_columns = ['ALL SEXES (RURAL)', 'MALE (RURAL)', 'FEMALE (RURAL)', 'TRANSGENDER (RURAL)']
urban_columns = ['ALL SEXES (URBAN)', 'MALE (URBAN)', 'FEMALE (URBAN)', 'TRANSGENDER (URBAN)']
total_rural_population = df[rural_columns].sum().values
total_urban_population = df[urban_columns].sum().values
plt.figure(figsize=(10, 6))
plt.bar(['Rural All Sexes', 'Rural Male', 'Rural Female', 'Rural Transgender'], total_rural_population, color='blue', alpha=0.7, label='Rural')
plt.bar(['Urban All Sexes', 'Urban Male', 'Urban Female', 'Urban Transgender'], total_urban_population, color='green', alpha=0.7, label='Urban')

plt.xlabel('Population Category')
plt.ylabel('Population')
plt.title('Population Distribution in Rural and Urban Areas')
plt.legend()
plt.show()
````

##  Question 6:Which province or division has the highest population growth rate in rural and urban areas?

````python

population_2017_rural_col = 'ALL SEXES (RURAL)'  
population_1998_rural_col = 'POPULATION 1998 (RURAL)'  
population_2017_urban_col = 'ALL SEXES (URBAN)'  
population_1998_urban_col = 'POPULATION 1998 (URBAN)'  


df['POPULATION GROWTH RATE (RURAL)'] = (df[population_2017_rural_col] - df[population_1998_rural_col]) / df[population_1998_rural_col] * 100
df['POPULATION GROWTH RATE (URBAN)'] = (df[population_2017_urban_col] - df[population_1998_urban_col]) / df[population_1998_urban_col] * 100


max_growth_rate_rural = df.loc[df['POPULATION GROWTH RATE (RURAL)'].idxmax()]
max_growth_rate_urban = df.loc[df['POPULATION GROWTH RATE (URBAN)'].idxmax()]

print("Province/Division with the highest population growth rate in rural areas:")
print(max_growth_rate_rural[['PROVINCE', 'DIVISION', 'POPULATION GROWTH RATE (RURAL)']])

print("\nProvince/Division with the highest population growth rate in urban areas:")
print(max_growth_rate_urban[['PROVINCE', 'DIVISION', 'POPULATION GROWTH RATE (URBAN)']])
````

##  Question 7: What is the gender distribution in rural and urban areas?

````python
total_male_rural = df['MALE (RURAL)'].sum()
total_female_rural = df['FEMALE (RURAL)'].sum()

total_male_urban = df['MALE (URBAN)'].sum()
total_female_urban = df['FEMALE (URBAN)'].sum()
print("Gender distribution in rural areas:")
print("Total Males:", total_male_rural)
print("Total Females:", total_female_rural)

print("\nGender distribution in urban areas:")
print("Total Males:", total_male_urban)
print("Total Females:", total_female_urban)
````

##  Question 8:Which districts have the highest and lowest population densities?

````python
df['POPULATION DENSITY'] = df['ALL SEXES (RURAL)'] / df['AREA (sq.km)']

district_with_highest_density = df.loc[df['POPULATION DENSITY'].idxmax()]
district_with_lowest_density = df.loc[df['POPULATION DENSITY'].idxmin()]
print("District with the highest population density:")
print(district_with_highest_density[['DISTRICT', 'POPULATION DENSITY']])

print("\nDistrict with the lowest population density:")
print(district_with_lowest_density[['DISTRICT', 'POPULATION DENSITY']])
````

##  Question 9:What are the trends in household sizes in different regions?


````python
avg_household_size_rural = df.groupby(df.index)['AVG HOUSEHOLD SIZE (RURAL)'].mean()
avg_household_size_urban = df.groupby(df.index)['AVG HOUSEHOLD SIZE (URBAN)'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_household_size_rural.index, avg_household_size_rural, label='Rural')
plt.plot(avg_household_size_urban.index, avg_household_size_urban, label='Urban')

plt.xlabel('Index')
plt.ylabel('Average Household Size')
plt.title('Trends in Household Sizes in Rural and Urban Areas')
plt.legend()
plt.grid(True)
plt.show()
````

##  Question 10:How does the population distribution between rural and urban areas compare within each province?

````python

population_by_province = df.groupby('PROVINCE')[['ALL SEXES (RURAL)', 'ALL SEXES (URBAN)']].sum()

provinces = population_by_province.index
population_rural = population_by_province['ALL SEXES (RURAL)']
population_urban = population_by_province['ALL SEXES (URBAN)']
bar_width = 0.35
r1 = np.arange(len(provinces))
r2 = [x + bar_width for x in r1]
plt.figure(figsize=(12, 6))
plt.bar(r1, population_rural, color='b', width=bar_width, edgecolor='grey', label='Rural')
plt.bar(r2, population_urban, color='r', width=bar_width, edgecolor='grey', label='Urban')
plt.xlabel('Province', fontweight='bold')
plt.ylabel('Population', fontweight='bold')
plt.xticks([r + bar_width/2 for r in range(len(provinces))], provinces, rotation=45)
plt.title('Population Distribution between Rural and Urban Areas by Province', fontweight='bold')

plt.legend()
plt.tight_layout()
plt.show()
````
